```python
import re
import numpy as np
compiled_regex = re.compile(r"score rank-[1-5]-bg")
phone_name=[]
cost=[]
spec_score=[]
a1, b1, c1, d1, e1, f1, g1, h1 = [], [], [], [], [], [], [], []
for i in container:
    phone_name.append(i.find('h2').text)
    cost.append(i.find('span',{'class':'price'}).text)
    spec_score.append(i.find('div',{'class':compiled_regex}).find('b').text)
    try:
        a1.append(i.text.replace('\u2009',' '))
    except:
        a1.append(np.nan)
    try:
        b1.append(i.text.replace('\u2009',' '))
    except:
        b1.append(np.nan)
    try:
        c1.append(i.text.replace('\u2009',' '))
    except:
        c1.append(np.nan)
    try:
        d1.append(i.text.replace('\u2009',' '))
    except:
        d1.append(np.nan)
    try:
        e1.append(i.text.replace('\u2009',' '))
    except:
        e1.append(np.nan)
    try:
        f1.append(i.text.replace('\u2009',' '))
    except:
        f1.append(np.nan)
    try:
        g1.append(i.text.replace('\u2009',' '))
    except:
        g1.append(np.nan) 
    try:
        h1.append(i.text.replace('\u2009',' '))
    except:
        h1.append(np.nan)
    
```

```python
import pandas as pd 
anew1=pd.DataFrame({
    'phone_name':phone_name,
    'cost':cost,
    'spec_score':spec_score,
    'Sim_specification':a1,
    'Processor_specification':b1,
    'memory_specification':c1,
    'battery_specification':d1,
    'display_specification':e1,
    'camera_specification':f1,
    'android_specification':g1,
    'fm_radio_specification':h1
})
```
